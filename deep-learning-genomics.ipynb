```python
#q1


!pip -q install tensorflow==2.15.0 scikit-learn==1.3.2 matplotlib==3.8.0 numpy==1.24.4

import os, random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, precision_recall_fscore_support,
                             roc_auc_score, confusion_matrix, ConfusionMatrixDisplay)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# hard seeds
SEED = 42
os.environ["PYTHONHASHSEED"] = str(SEED)
os.environ["TF_DETERMINISTIC_OPS"] = "1"
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

#downlaoding data from github
import urllib.request, pathlib

RAW_SEQ_URL = "https://raw.githubusercontent.com/abidlabs/deep-learning-genomics-primer/master/sequences.txt"
RAW_LAB_URL = "https://raw.githubusercontent.com/abidlabs/deep-learning-genomics-primer/master/labels.txt"
DATA_DIR = pathlib.Path("genomics_data")
DATA_DIR.mkdir(exist_ok=True)
SEQ_PATH = DATA_DIR / "sequences.txt"
LAB_PATH = DATA_DIR / "labels.txt"

def maybe_download(url, path):
    if not path.exists():
        print(f"Downloading {path.name} ...")
        urllib.request.urlretrieve(url, path)
    else:
        print(f"Found {path.name}")

maybe_download(RAW_SEQ_URL, SEQ_PATH)
maybe_download(RAW_LAB_URL, LAB_PATH)


with open(SEQ_PATH, "r") as f:
    sequences = [line.strip().upper() for line in f if line.strip()]

labels = np.loadtxt(LAB_PATH, dtype=int)
assert len(sequences) == len(labels), "Mismatch: sequences and labels length differ."


alphabet = set("ACGT")
mask_valid = np.array([set(s).issubset(alphabet) for s in sequences])
sequences = [s for s, ok in zip(sequences, mask_valid) if ok]
labels = labels[mask_valid]
print(f"Total valid examples: {len(sequences)}")


lengths = np.array([len(s) for s in sequences])
L = int(np.median(lengths))  # target length from median

def center_trim(s, L):
    if len(s) == L:
        return s
    if len(s) > L:
        start = (len(s) - L) // 2
        return s[start:start+L]
    return None

trimmed = []
keep_idx = []
for i, s in enumerate(sequences):
    t = center_trim(s, L)
    if t is not None:
        trimmed.append(t)
        keep_idx.append(i)
sequences = trimmed
labels = labels[np.array(keep_idx)]
print(f"Using fixed length L={L}. Kept {len(sequences)} samples after trim/filter.")

# balanced subsets up to 10k pos + 10k neg
pos_idx = np.where(labels == 1)[0]
neg_idx = np.where(labels == 0)[0]
n_pos = min(10_000, len(pos_idx))
n_neg = min(10_000, len(neg_idx))
if n_pos == 0 or n_neg == 0:
    raise ValueError("Dataset does not contain both positive and negative examples after filtering.")
rng = np.random.default_rng(SEED)
pos_sel = rng.choice(pos_idx, size=n_pos, replace=False)
neg_sel = rng.choice(neg_idx, size=n_neg, replace=False)
sel = np.concatenate([pos_sel, neg_sel])
rng.shuffle(sel)
sequences = [sequences[i] for i in sel]
labels = labels[sel].astype(int)

print(f"Subset sizes → pos: {np.sum(labels==1)}, neg: {np.sum(labels==0)}, total: {len(labels)}")

# train/val/test
X_train, X_tmp, y_train, y_tmp = train_test_split(
    sequences, labels, test_size=0.30, random_state=SEED, stratify=labels
)
X_val, X_test, y_val, y_test = train_test_split(
    X_tmp, y_tmp, test_size=0.50, random_state=SEED, stratify=y_tmp
)
print(f"Splits → train: {len(X_train)}, val: {len(X_val)}, test: {len(X_test)}")


base2idx = {'A':0, 'C':1, 'G':2, 'T':3}

def encode_label(seqs):
    # A->0, C->1, G->2, T->3 -> shape (N, L) int
    arr = np.zeros((len(seqs), L), dtype=np.int32)
    for i, s in enumerate(seqs):
        arr[i] = [base2idx[b] for b in s]
    return arr

def encode_onehot(seqs):

    out = np.zeros((len(seqs), L, 4), dtype=np.float32)
    for i, s in enumerate(seqs):
        idxs = [base2idx[b] for b in s]
        out[i, np.arange(L), idxs] = 1.0
    return out.reshape(len(seqs), L*4)


def fcgr_image_for_sequence(seq, k=4):
    bins = 2**k
    corner = {
        'A': np.array([0.0, 0.0]),
        'C': np.array([0.0, 1.0]),
        'G': np.array([1.0, 1.0]),
        'T': np.array([1.0, 0.0]),
    }
    p = np.array([0.5, 0.5], dtype=np.float64)
    points = np.empty((len(seq), 2), dtype=np.float64)
    for i, b in enumerate(seq):
        p = 0.5*(p + corner[b])
        points[i] = p
    H, xedges, yedges = np.histogram2d(points[:,0], points[:,1], bins=bins, range=[[0,1],[0,1]])

    if H.sum() > 0:
        H = H / H.sum()
    return H.astype(np.float32)

def encode_fcgr(seqs, k=4):
    imgs = np.stack([fcgr_image_for_sequence(s, k=k) for s in seqs], axis=0)
    return imgs.reshape(len(seqs), -1)


Xtr_lab, Xva_lab, Xte_lab = encode_label(X_train), encode_label(X_val), encode_label(X_test)
Xtr_one, Xva_one, Xte_one = encode_onehot(X_train), encode_onehot(X_val), encode_onehot(X_test)

Xtr_fcg, Xva_fcg, Xte_fcg = encode_fcgr(X_train, k=4), encode_fcgr(X_val, k=4), encode_fcgr(X_test, k=4)

print("Shapes:")
print(" Label-encoded:", Xtr_lab.shape, Xva_lab.shape, Xte_lab.shape)
print(" One-hot:", Xtr_one.shape, Xva_one.shape, Xte_one.shape)
print(" FCGR:", Xtr_fcg.shape, Xva_fcg.shape, Xte_fcg.shape)


def build_ann_label(input_len=L, vocab=4, embed_dim=8, hidden=64, dropout=0.2):
    inp = keras.Input(shape=(input_len,), dtype="int32")
    x = layers.Embedding(input_dim=vocab, output_dim=embed_dim, input_length=input_len)(inp)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(hidden, activation="relu")(x)
    x = layers.Dropout(dropout)(x)
    out = layers.Dense(1, activation="sigmoid")(x)
    model = keras.Model(inp, out)
    model.compile(optimizer="adam",
                  loss="binary_crossentropy",
                  metrics=[keras.metrics.AUC(name="auc"), "accuracy"])
    return model

def build_ann_vector(input_dim, hidden=64, dropout=0.2):
    inp = keras.Input(shape=(input_dim,), dtype="float32")
    x = layers.Dense(hidden, activation="relu")(inp)
    x = layers.Dropout(dropout)(x)
    x = layers.Dense(hidden//2, activation="relu")(x)
    out = layers.Dense(1, activation="sigmoid")(x)
    model = keras.Model(inp, out)
    model.compile(optimizer="adam",
                  loss="binary_crossentropy",
                  metrics=[keras.metrics.AUC(name="auc"), "accuracy"])
    return model


BATCH = 128
EPOCHS = 12
callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_auc", mode="max", patience=3, restore_best_weights=True)
]

def train_and_report(name, model, Xtr, ytr, Xva, yva, Xte, yte, primary_metric="auc"):
    print(f"\n===== {name}: training =====")
    hist = model.fit(
        Xtr, ytr, validation_data=(Xva, yva),
        epochs=EPOCHS, batch_size=BATCH, callbacks=callbacks, verbose=2
    )

    fig, ax = plt.subplots(1, 2, figsize=(12,4))
    ax[0].plot(hist.history["loss"], label="train")
    ax[0].plot(hist.history["val_loss"], label="val")
    ax[0].set_title(f"{name} — Loss")
    ax[0].set_xlabel("Epoch"); ax[0].set_ylabel("Binary Cross-Entropy"); ax[0].legend()

    ax[1].plot(hist.history[primary_metric], label="train")
    ax[1].plot(hist.history[f"val_{primary_metric}"], label="val")
    ax[1].set_title(f"{name} — {primary_metric.upper()}")
    ax[1].set_xlabel("Epoch"); ax[1].set_ylabel(primary_metric.upper()); ax[1].legend()
    plt.show()


    y_prob = model.predict(Xte, batch_size=BATCH, verbose=0).ravel()
    y_pred = (y_prob >= 0.5).astype(int)
    acc = accuracy_score(yte, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(yte, y_pred, average="binary", zero_division=0)
    try:
        auc = roc_auc_score(yte, y_prob)
    except ValueError:
        auc = float("nan")

    print(f"{name} — Test metrics")
    print(f"  Accuracy:  {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall:    {rec:.4f}")
    print(f"  F1-score:  {f1:.4f}")
    print(f"  ROC-AUC:   {auc:.4f}")


    cm = confusion_matrix(yte, y_pred, labels=[0,1])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
    fig, ax = plt.subplots(figsize=(4.5,4))
    disp.plot(ax=ax, values_format="d", colorbar=False)
    ax.set_title(f"{name} — Test Confusion Matrix")
    plt.show()


    params = model.count_params()
    print(f"{name} — Total parameters: {params:,}")

    return {
        "history": hist.history,
        "metrics": {"accuracy": acc, "precision": prec, "recall": rec, "f1": f1, "roc_auc": auc},
        "params": params
    }


model_label = build_ann_label(input_len=L, embed_dim=8, hidden=64, dropout=0.2)
res_label = train_and_report("Label Encoding (Embedding+ANN)", model_label,
                             Xtr_lab, y_train, Xva_lab, y_val, Xte_lab, y_test)

model_onehot = build_ann_vector(input_dim=Xtr_one.shape[1], hidden=64, dropout=0.2)
res_onehot = train_and_report("One-Hot (Flattened) + ANN", model_onehot,
                              Xtr_one, y_train, Xva_one, y_val, Xte_one, y_test)

model_fcgr = build_ann_vector(input_dim=Xtr_fcg.shape[1], hidden=64, dropout=0.2)
res_fcgr = train_and_report("FCGR (k=4, 16x16) + ANN", model_fcgr,
                            Xtr_fcg, y_train, Xva_fcg, y_val, Xte_fcg, y_test)


import pandas as pd
summary = pd.DataFrame.from_dict({
    "Encoding": ["Label (Embedding)", "One-Hot", "FCGR (k=4)"],
    "Params": [res_label["params"], res_onehot["params"], res_fcgr["params"]],
    "Test Accuracy": [res_label["metrics"]["accuracy"], res_onehot["metrics"]["accuracy"], res_fcgr["metrics"]["accuracy"]],
    "Test Precision": [res_label["metrics"]["precision"], res_onehot["metrics"]["precision"], res_fcgr["metrics"]["precision"]],
    "Test Recall": [res_label["metrics"]["recall"], res_onehot["metrics"]["recall"], res_fcgr["metrics"]["recall"]],
    "Test F1": [res_label["metrics"]["f1"], res_onehot["metrics"]["f1"], res_fcgr["metrics"]["f1"]],
    "Test ROC-AUC": [res_label["metrics"]["roc_auc"], res_onehot["metrics"]["roc_auc"], res_fcgr["metrics"]["roc_auc"]],
})
print("\n=== Q1 Summary (Test Set) ===")
display(summary.round(4))



```

# Q1 — Turning Nucleotides into Numbers for ANN

## Task Description
This assignment performs a binary sequence classification task based on the dataset used in the notebook “A Primer on Deep Learning in Genomics.” Each DNA sequence is labeled as either 0 or 1, indicating whether a transcription factor (TF) is predicted to bind that sequence. The goal is to encode DNA (A, C, G, T) into numeric representations suitable for neural networks and train small ANNs on each encoding.

## Dataset Slice Used
**Source:** `sequences.txt` and `labels.txt` from Abid’s deep learning genomics GitHub

- Each sample: one nucleotide sequence (string)
- Each label: binary (0/1)

All sequences were validated to ensure they contained only A, C, G, and T.

## Modifications and Preprocessing Steps
To ensure fairness across encodings:

- Filtered sequences containing characters outside {A, C, G, T}
- Computed the median sequence length and center-trimmed sequences longer than this length
- Dropped sequences shorter than the median to maintain a fixed length
- Formed a balanced subset (up to 10,000 positive and 10,000 negative examples)
- Shuffled using a fixed random seed

After preprocessing, a balanced subset suitable for training small neural networks was obtained.

## Reproducibility
To ensure deterministic behavior:

- `random.seed`, `numpy.random.seed`, and `tf.random.set_seed` were set
- `PYTHONHASHSEED` was fixed
- TensorFlow deterministic operations were enabled

## Train, Validation, and Test Splitting
A stratified split was used to preserve label ratios:

- 70% training
- 15% validation
- 15% testing

This prevents label distribution bias.

## Evaluation Metrics
For each encoding strategy, the following are reported:

- Accuracy
- Precision
- Recall
- F1-score
- ROC-AUC (binary)
- Test confusion matrix

These metrics evaluate performance, threshold sensitivity, and class bias.

## Experiment Structure
Each encoding experiment:

- Trains and prints loss/AUC curves
- Computes test metrics
- Displays a confusion matrix
- Reports total parameter count for fairness across architectures

## Encoding Strategies Tested

### 1. Label Encoding + Embedding
- Converts A → 0, C → 1, G → 2, T → 3
- Uses a learnable embedding lookup
- Compact and memory-efficient

### 2. One-Hot Encoding (Flattened)
- Represents each nucleotide as a 4-vector
- Preserves categorical nature without artificial ordering

### 3. FCGR (Chaos Game Representation)
- Converts sequence into a 2D frequency grid
- Captures compositional and k-mer patterns

All three were fed into small ANNs of comparable capacity.

## Training Plots
For each encoding, the following were plotted:

- Binary cross-entropy loss (to visualize optimization convergence)
- ROC-AUC (to measure ranking ability independent of threshold)

Training versus validation curves allowed diagnosis of under- or over-fitting.

## Final Test Results
All metrics were computed on the held-out test set. The confusion matrix summarized:

- True positives (correct bound predictions)
- True negatives (correct non-bound predictions)
- False positives and false negatives

## Pros and Cons Summary
A short evaluation was provided comparing:

- Parameter footprint
- Ability to capture positional patterns
- Biological interpretability
- Memory cost

## Conclusion of Q1
This question demonstrated how the same genomic data can be represented in three fundamentally different numeric forms, and how those choices affect:

- Model efficiency
- Generalization behavior
- Representational richness

These differences are important in genomics, where local motifs, composition, and positional context strongly influence prediction quality.


## Short Explanation of Results

**Label Encoding**
- Performed well with a low parameter count.
- Achieved a reasonable balance of accuracy, recall, and AUC.
- The embedding layer helped the model learn meaningful representations despite the ordinal nature of labels.

**One-Hot Encoding**
- Increased dimensionality and parameter count.
- Delivered competitive but not superior performance.
- Preserved nucleotide identity faithfully, but required more model capacity to generalize effectively.

**FCGR (Chaos Game Representation)**
- Performed the best, reaching near-perfect generalization in this setup.
- Captured higher-order k-mer and compositional patterns using a compact 2D histogram.
- Learned global genomic signatures effectively.
- However, discarded strict positional information and may overfit on smaller datasets without proper regularization.


## Comparison of Encoding Strategies

### 1) Label Encoding (A=0, C=1, G=2, T=3) + Embedding

**Pros**
- Very compact input (L integers) → few parameters with a small embedding.
- Allows the model to learn distributed representations of nucleotides.
- Fast to train; small memory footprint.

**Cons**
- Introduces an implicit ordinal relationship (0 < 1 < 2 < 3) unless using an embedding layer (here, we do).
- May underperform if higher-order interactions (k-mers) are critical and the model is shallow.

---

### 2) One-Hot Encoding (L × 4), flattened to a vector

**Pros**
- Lossless representation with no artificial ordinal structure.
- Simple, interpretable, and often a strong baseline.

**Cons**
- 4× more features than label encoding → larger input dimensionality.
- Requires more parameters or stronger regularization to avoid overfitting.

---

### 3) FCGR (Chaos Game Representation) with k=4 → 16×16 image flattened

**Pros**
- Encodes compositional and higher-order k-mer patterns.
- Compact (256 features) and robust to small positional shifts or noise.

**Cons**
- Loses exact positional information along the sequence.
- Choice of k impacts bias–variance:
  - Too small → loses detail
  - Too large → very sparse counts

---

### Notes
- All three models used comparable-capacity ANNs (same hidden widths), and parameter counts were reported to ensure fairness.
- Primary comparison metric is ROC-AUC; accuracy, precision, recall, and F1-score are also reported as required.


```python
#utility functions for q2

def make_splits(X, y, seed=SEED):
    """Stratified 70/15/15 split. Returns (Xtr, Xva, Xte, ytr, yva, yte)."""
    Xtr, Xtmp, ytr, ytmp = train_test_split(
        X, y, test_size=0.30, random_state=seed, stratify=y
    )
    Xva, Xte, yva, yte = train_test_split(
        Xtmp, ytmp, test_size=0.50, random_state=seed, stratify=ytmp
    )
    return Xtr, Xva, Xte, ytr, yva, yte


def report_metrics(y_true, y_prob, title="Test"):
    """Print Acc/Prec/Rec/F1/ROC-AUC and draw confusion matrix."""
    y_pred = (y_prob >= 0.5).astype(int)

    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average="binary", zero_division=0
    )

    try:
        auc = roc_auc_score(y_true, y_prob)
    except ValueError:
        auc = float("nan")

    print(f"{title} — Metrics")
    print(f"  Accuracy : {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall   : {rec:.4f}")
    print(f"  F1-score : {f1:.4f}")
    print(f"  ROC-AUC  : {auc:.4f}")

    cm = confusion_matrix(y_true, y_pred, labels=[0,1])
    ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1]).plot()


def plot_curves(history, title_prefix="ANN"):
    """Plot training/validation loss and AUC curves."""
    h = history.history
    fig, ax = plt.subplots(1, 2, figsize=(12,4))

    ax[0].plot(h["loss"], label="train")
    ax[0].plot(h["val_loss"], label="val")
    ax[0].set_title(f"{title_prefix} — Loss"); ax[0].legend()

    ax[1].plot(h["auc"], label="train")
    ax[1].plot(h["val_auc"], label="val")
    ax[1].set_title(f"{title_prefix} — AUC"); ax[1].legend()

    plt.show()


def run_model(model, Xtr, ytr, Xva, yva, Xte, yte, title):
    hist = model.fit(
        Xtr, ytr,
        validation_data=(Xva, yva),
        epochs=EPOCHS,
        batch_size=BATCH,
        callbacks=callbacks,
        verbose=2
    )
    plot_curves(hist, title_prefix=title)

    y_prob = model.predict(Xte).ravel()
    metrics = report_metrics(yte, y_prob, title)

    print(f"{title} — params: {model.count_params():,}")
    return metrics, model.count_params()

```

## 5. Why Accuracy Alone Is Not Sufficient

Although accuracy is intuitive, it has several limitations:

1. **It ignores error type.**
   Failing to detect a true positive binding site can be more biologically costly than a false positive.

2. **It does not capture trade-offs.**
   Precision vs. recall trade-offs are critical in genomics (sensitivity is often prioritized).

3. **Threshold dependency.**
   Accuracy only evaluates predictions at a single probability cutoff (0.5), while data distributions vary.

4. **It ignores ranking quality.**
   ROC-AUC scores the model’s ability to order outputs correctly across all possible thresholds.

5. **It masks class imbalance effects.**
   Even mild imbalance can inflate accuracy deceptively.

For these reasons, genomics tasks are typically evaluated with:
- Precision, Recall
- F1-Score
- ROC-AUC

which the assignment requires and we report.

```python
#q3 a

def build_deep_ann(input_len=L, hidden=64, dropout=0.2):
    inp = keras.Input(shape=(input_len,), dtype="int32")
    x = layers.Embedding(input_dim=4, output_dim=8)(inp)
    x = layers.GlobalAveragePooling1D()(x)

    for _ in range(6):  # original + 5 extra
        x = layers.Dense(hidden, activation="relu")(x)

    x = layers.Dropout(dropout)(x)
    out = layers.Dense(1, activation="sigmoid")(x)
    model = keras.Model(inp, out)
    model.compile(optimizer="adam", loss="binary_crossentropy",
                  metrics=[keras.metrics.AUC(name="auc"), "accuracy"])
    return model

deep_model = build_deep_ann()
deep_results, deep_params = run_model(deep_model, Xtr, ytr, Xva, yva, Xte, yte,
                                      "Deep ANN (6 hidden layers)")
```

## Q4-A — Concept: How a 1D CNN Differs From an ANN for Sequence Classification

For genomic sequence classification, Artificial Neural Networks (ANNs) treat the entire input as a flat vector. This means the network does not inherently understand local sequence patterns, such as short binding motifs or k-mer signatures, unless they are learned indirectly through dense connections. Every input position interacts with every neuron, making ANNs parameter-heavy and less spatially aware.

In contrast, a 1D Convolutional Neural Network (CNN) introduces inductive bias toward local, position-dependent features by sliding convolutional filters across the sequence. This makes CNNs well-suited for biological sequence tasks where small, contiguous nucleotide patterns are highly discriminative.

---

### 1. Kernel Size (Filter Width)
The kernel size determines how many adjacent bases are processed at once.

- Kernel size = 5 → captures short regulatory motifs
- Kernel size = 9 → captures slightly longer dependencies

Small kernels specialize in short patterns; larger kernels detect broader context. This localized receptive field allows CNNs to automatically learn biologically meaningful motif-like structures.

---

### 2. Stride
The stride controls how far the filter moves after each step.

- Stride = 1 → preserves resolution and detects fine motifs
- Stride > 1 → performs implicit down-sampling

In genomics, stride = 1 is common to avoid skipping informative k-mers.

---

### 3. Pooling (e.g., Max Pooling)
Pooling reduces spatial dimensionality by summarizing features across neighborhoods.

- Max pooling keeps the strongest activation
- Provides translation invariance (motif anywhere is acceptable)
- Reduces overfitting and memory usage

Genomic motifs can shift slightly without changing biological meaning; pooling supports this.

---

### 4. Non-Linear Activation
After each convolution, non-linear activations (e.g., ReLU):

- Increase feature selectivity
- Introduce sparsity
- Break linearity

Without them, stacked layers collapse to a single linear filter. ReLU preserves gradient strength and avoids saturation, enabling deeper networks.

---

### 5. Weight Sharing
Convolutional filters reuse the same weights as they slide:

- Dramatically fewer parameters than dense ANNs
- Translational robustness
- Efficient learning of repeated motifs

This is ideal for biological data where patterns recur across positions.

---

### 6. Biological Interpretation Advantage
Learned CNN filters often correspond to:

- Transcription factor binding motifs
- Short consensus sequences
- Regulatory subsequences

Thus, CNNs improve interpretability in addition to accuracy.

---

### 7. Summary Difference: ANN vs. 1D CNN

| Property                  | ANN            | 1D CNN                         |
|---------------------------|----------------|--------------------------------|
| Local pattern detection   | Weak           | Strong                         |
| Positional invariance     | None           | Pooling provides invariance    |
| Parameter count           | High           | Reduced via weight sharing     |
| Motif modeling            | Indirect       | Explicit                       |
| Interpretability          | Lower          | Higher (filter-motif mapping)  |

---

### Final Conceptual Takeaway
A 1D CNN improves performance for sequence tasks because it:

- Detects local patterns through convolutional kernels
- Preserves motif presence regardless of position via pooling
- Uses non-linear activations to stack expressive filters
- Shares weights to reduce parameters and improve generalization

This aligns with how real genomic regulatory motifs behave, making CNNs a natural architectural fit for sequence-based bioinformatics.

- Is sufficiently expressive to combine motif-level features
- Avoids unnecessary parameter inflation

Instead of flattening, global max pooling:
- Aggregates the strongest activation from each filter
- Answers: “Did this motif appear anywhere in the sequence?”
- Drastically reduces parameter count
- Makes the dense head more efficient and less prone to overfitting

- Produces probability scores between 0 and 1
